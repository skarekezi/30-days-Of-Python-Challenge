```python
#Day 11:Functions
# Simple function
def greet(name):
    return f"Hello, {name}!"

print(greet("Orlane"))

# Function with default
def multiply(a, b=2):
    return a * b

print(multiply(5))
print(multiply(5,3))

# Multiple returns
def operations(a,b):
    return a+b, a-b, a*b

sum_val, diff, product = operations(10,5)
print(sum_val, diff, product)

# Variable-length args
def sum_all(*args):
    return sum(args)
print(sum_all(1,2,3,4))
```

```python
#Day 13: List comprehension
nums = [1,2,3,4,5]

# Simple comprehension
squared = [x**2 for x in nums]
print(squared)

# Conditional
evens = [x for x in nums if x%2==0]
print(evens)

# Nested comprehension
matrix = [[1,2],[3,4]]
flat = [n for row in matrix for n in row]
print(flat)

# Mapping
names = ["Orlane","Amina"]
upper_names = [name.upper() for name in names]
print(upper_names)

# With condition and transformation
doubles = [x*2 for x in nums if x>2]
print(doubles)
```

```python
#Day 14: Higher order Functions
from functools import reduce

nums = [1,2,3,4]

# map
squared = list(map(lambda x: x**2, nums))
print(squared)

# filter
greater_than_two = list(filter(lambda x: x>2, nums))
print(greater_than_two)

# reduce
total = reduce(lambda a,b: a+b, nums)
print(total)

# Combined
result = reduce(lambda a,b: a*b, filter(lambda x:x%2==0, nums))
print("Product of evens:", result)
```

```python
#Day15: Python Type erros
# 1. Type conversion
a = "10"
b = 5
print(int(a) + b)   # Converts string to int

# 2. Float to int
x = 3.7
y = 2
print(int(x) + y)   # 3 + 2 = 5

# 3. String concatenation
first_name = "Orlane"
last_name = "Karekezi"
full_name = first_name + " " + last_name
print(full_name)

# 4. Invalid operations causing errors
try:
    print("10" + 5)  # TypeError
except TypeError:
    print("Cannot add string and int directly!")

# 5. List and string type issue
numbers = [1,2,3]
try:
    print(numbers + "hello")  # TypeError
except TypeError:
    print("Cannot concatenate list and string")

# 6. Boolean type as int
is_student = True
print("Is student:", is_student + 1)  # True is treated as 1

# 7. Converting non-numeric string to int
try:
    print(int("Python"))
except ValueError:
    print("Cannot convert non-numeric string to int")

# 8. Multiple type conversion
num_str = "12.5"
num_float = float(num_str)
num_int = int(num_float)
print("Float:", num_float, "Int:", num_int)
```

```python
#Day 16:Date & time
from datetime import datetime, timedelta

now = datetime.now()
print("Now:", now)

birthday = datetime(2005,6,15)
print("Birthday:", birthday)

# Difference
print("Days lived:", (now - birthday).days)

# Future date
future_event = now + timedelta(days=100)
print("100 days later:", future_event)

# Formatting
print(now.strftime("%A, %d %B %Y"))

# Parsing from string
date_str = "2025-12-25"
parsed_date = datetime.strptime(date_str, "%Y-%m-%d")
print(parsed_date)

```

```python
#Day 17: Exception Handling
# Try/except
try:
    value = int(input("Enter a number: "))
    print(10/value)
except ValueError:
    print("Invalid number")
except ZeroDivisionError:
    print("Cannot divide by zero")
finally:
    print("Done")

# Raising exception
def check_positive(num):
    if num < 0:
        raise ValueError("Number must be positive")
check_positive(10)

# Catch multiple exceptions
try:
    nums = [1,2,3]
    print(nums[5])
except (IndexError, KeyError) as e:
    print("Error:", e)
```

```python
#Day18:Regular Expression
import re

text = "Contact: alice@mail.com, bob@work.org, call 123-456-7890 or 987-654-3210"

# Find all emails
emails = re.findall(r"\b\S+@\S+\.\S+\b", text)
print("Emails:", emails)

# Find all phone numbers
phones = re.findall(r"\d{3}-\d{3}-\d{4}", text)
print("Phones:", phones)

# Search for pattern
match = re.search(r"alice@\S+", text)
if match:
    print("Found:", match.group())

# Replace emails with [hidden]
masked = re.sub(r"\b\S+@\S+\.\S+\b", "[hidden]", text)
print(masked)

# Split text by spaces
words = re.split(r"\s", text)
print("Words:", words)

# Check pattern
pattern = r"^\d{3}-\d{3}-\d{4}$"
print("Valid phone?", bool(re.match(pattern, "123-456-7890")))
```

```python
#Day 21:Classes and Objects
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def greet(self):
        return f"Hello, {self.name}!"

# Create objects
p1 = Person("Alice", 25)
p2 = Person("Bob", 30)

print(p1.greet())
print(p2.greet())

# Add attribute dynamically
p1.city = "Nairobi"
print(p1.name, "lives in", p1.city)

# Class method
class Student:
    school = "USIU"
    def __init__(self, name):
        self.name = name
    def info(self):
        return f"{self.name} studies at {Student.school}"

s = Student("Orlane")
print(s.info())
```

```python
#Day 22:Web Scrapping
import requests
from bs4 import BeautifulSoup

url = "https://books.toscrape.com/catalogue/a-light-in-the-attic_1000/index.html"
res = requests.get(url)
soup = BeautifulSoup(res.text, "html.parser")

# Extract book title
title = soup.select_one("h1").get_text()
print("Title:", title)

# Extract book description (paragraph)
description = soup.select_one("#product_description ~ p").get_text()
print("Description:", description)

# Extract price
price = soup.select_one(".price_color").get_text()
print("Price:", price)
```

```python
#Day 24:Statistics
import matplotlib.pyplot as plt
from statistics import mean, median, mode, variance, stdev

# Sample data (list of numbers)
ages = [22, 25, 30, 28, 22, 30, 25]

# Statistics
print("Mean:", mean(ages))
print("Median:", median(ages))
print("Mode:", mode(ages))
print("Variance:", variance(ages))
print("Standard Deviation:", stdev(ages))

# Graph: Bar chart of frequency
from collections import Counter
counts = Counter(ages)
plt.bar(counts.keys(), counts.values(), color='skyblue')
plt.title("Frequency of Ages")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
#Day 29: Building API
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# 1️⃣ Simulate API response (JSON-like data)
def get_student_scores():
    return [
        {"name": "Alice", "score": 85, "age": 25},
        {"name": "Bob", "score": 90, "age": 30},
        {"name": "Charlie", "score": 78, "age": 22},
        {"name": "David", "score": 88, "age": 28},
        {"name": "Eve", "score": 92, "age": 27}
    ]

# Call the “API”
students = get_student_scores()
print("API Response:", students)

# 2️⃣ Convert to DataFrame
df = pd.DataFrame(students)
print("\nDataFrame:\n", df)

# 3️⃣ Optional: Bar chart of Scores
df.plot(kind="bar", x="name", y="score", color="skyblue", title="Student Scores from API")
plt.ylabel("Score")
plt.show()

# 4️⃣ Optional: Pie chart for Score distribution
df.set_index("name")["score"].plot(kind="pie", autopct="%1.1f%%", title="Score Distribution")
plt.show()

# 5️⃣ Filter: Students with score > 85
high_scores = df[df["score"] > 85]
print("\nStudents scoring above 85:\n", high_scores)

# 6️⃣ Summary statistics
print("\nSummary Statistics:")
print(df.describe())
```
